RFE (Recursive Feature Elimination)

RFE (Eliminación Recursiva de Características) es una técnica de selección de características muy utilizada en minería de datos y aprendizaje automático para mejorar la precisión y eficiencia de los modelos. Esta técnica selecciona las características más importantes reduciendo gradualmente las menos relevantes hasta quedarse solo con aquellas que tienen mayor impacto en el rendimiento del modelo.



```python
!pip install -U -q PyDrive

import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
```

```python
# Cargar los archivos CSV directamente desde Google Drive
dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/dataset_inicial.csv')
```

```python
# Vista rápida de las estructuras de los archivos
print(dataset.head())
```

```python
X = dataset.drop('s04a_06', axis=1)
y = dataset['s04a_06']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
model = LogisticRegression(max_iter=1000)
rfe = RFE(estimator=model, n_features_to_select=6)
X_rfe = rfe.fit_transform(X_scaled, y)
model = LogisticRegression(max_iter=1000)
rfe = RFE(estimator=model, n_features_to_select=6)
X_rfe = rfe.fit_transform(X_scaled, y)
selected_features = X.columns[rfe.support_]
print("Características seleccionadas:", selected_features)
print("Ranking of features:", rfe.ranking_)

```

```python
# Define a list to store the scores for each number of features
scores = []

# Define the range of number of features to test
n_features_list = list(range(1, X.shape[1] + 1)) # [1, 2, 3, ..., total_features]

# Iterate over the range of number of features
for n_features in n_features_list:
    # Create and train the model with RFE
    model = LogisticRegression(max_iter=1000)
    rfe = RFE(estimator=model, n_features_to_select=n_features)
    X_rfe = rfe.fit_transform(X_scaled, y)

    # Calculate the cross-validation score and append to the list
    cv_scores = cross_val_score(model, X_rfe, y, cv=5) # Example: 5-fold cross-validation
    scores.append(cv_scores.mean())

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(n_features_list, scores, marker='o', color='b')
plt.xlabel("Número de Características Seleccionadas")
plt.ylabel("Precisión Promedio de Validación Cruzada")
plt.title("Rendimiento del Modelo vs Número de Características Seleccionadas (RFE)")
plt.grid(True)
plt.show()
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer # Import SimpleImputer
import pandas as pd
import numpy as np

# Assuming 'df' is your original DataFrame
X_cla = df.drop(columns=['s04a_06'])  # Assuming 's04a_06' is your target column
y_cla = df['s04a_06']

# Identify categorical and numerical features
categorical_features = X_cla.select_dtypes(include=['object']).columns.tolist()
numerical_features = X_cla.select_dtypes(exclude=['object']).columns.tolist()

# Create a ColumnTransformer to apply different preprocessing to different columns
# Include SimpleImputer to handle NaN values in numerical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', SimpleImputer(strategy='mean'), numerical_features),  # Impute missing values with mean for numerical features
        ('cat', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), categorical_features),  # One-hot encoding for categorical features
    ])

# Fit and transform the data
X_cla_processed = preprocessor.fit_transform(X_cla)

#Choose one of the Options
filter_condition = ~np.isnan(y_cla) # Create a filter condition for non-NaN values in y_cla
X_cla_processed = X_cla_processed[filter_condition]
y_cla = y_cla[filter_condition] #Filter X_cla_processed and y_cla using the condition


test = SelectKBest(score_func=chi2, k=6)
fit = test.fit(X_cla_processed, y_cla)

# Get the feature names after transformation
feature_names = numerical_features + preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_features).tolist()

# Print the results
np.set_printoptions(precision=3)
print(feature_names)
print(fit.scores_)
print([feature_names[i] for i in fit.get_support(indices=True)]) #Get selected feature names
```
